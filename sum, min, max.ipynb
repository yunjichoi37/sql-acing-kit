### **<span style="color:#FFAEC9">8. 물고기 종류 별 대어 찾기</span>**

1. 서브쿼리를 두 겹으로 하는 방식: group by + join

```python
-- 1. 타입별 가장 큰 길이
SELECT FISH_TYPE, MAX(LENGTH)
FROM FISH_INFO
GROUP BY FISH_TYPE


-- 2. 타입별 가장 큰 길이와 해당하는 물고기의 ID
SELECT I.ID, T.FISH_TYPE, T.LENGTH
FROM (
    SELECT FISH_TYPE, MAX(LENGTH) AS LENGTH
    FROM FISH_INFO
    GROUP BY FISH_TYPE
    ) AS T
JOIN FISH_INFO AS I 
    ON I.FISH_TYPE = T.FISH_TYPE
    AND I.LENGTH = T.LENGTH
ORDER BY I.ID ASC


-- 3. 타입별 가장 큰 길이와 해당하는 물고기의 이름
SELECT
    MT.ID AS ID, 
    NI.FISH_NAME AS FISH_NAME,
    MT.LENGTH AS LENGTH
FROM (
    SELECT I.ID, T.FISH_TYPE, T.LENGTH
    FROM (
        SELECT FISH_TYPE, MAX(LENGTH) AS LENGTH
        FROM FISH_INFO
        GROUP BY FISH_TYPE
        ) AS T
    JOIN FISH_INFO AS I 
        ON I.FISH_TYPE = T.FISH_TYPE
        AND I.LENGTH = T.LENGTH
) AS MT
JOIN FISH_NAME_INFO AS NI
ON NI.FISH_TYPE = MT.FISH_TYPE
ORDER BY MT.ID ASC
```
