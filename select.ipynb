### 3. 과일로 만든 아이스크림 고르기

```python
SELECT F.FLAVOR
FROM FIRST_HALF AS F, ICECREAM_INFO AS I
WHERE F.TOTAL_ORDER > 3000
    AND I.INGREDIENT_TYPE = 'fruit_based'
    AND F.FLAVOR = I.FLAVOR
ORDER BY F.TOTAL_ORDER DESC
```

```python
SELECT F.FLAVOR
FROM FIRST_HALF AS F
JOIN ICECREAM_INFO AS I ON F.FLAVOR = I.FLAVOR
WHERE F.TOTAL_ORDER > 3000
    AND I.INGREDIENT_TYPE = 'fruit_based'
ORDER BY F.TOTAL_ORDER DESC
```

### 5. 12세 이하인 여자 환자 목록 출력하기

### 10. 서울에 위치한 식당 목록 출력하기

```python
SELECT I.REST_ID, I.REST_NAME, I.FOOD_TYPE, I.FAVORITES, I.ADDRESS, ROUND(AVG(R.REVIEW_SCORE), 2) AS SCORE
FROM REST_INFO AS I
JOIN REST_REVIEW AS R ON I.REST_ID = R.REST_ID
WHERE I.ADDRESS LIKE '서울%'
GROUP BY I.REST_ID
ORDER BY SCORE DESC, I.FAVORITES DESC
```

### 11. 재구매가 일어난 상품과 회원 리스트 구하기

```python
SELECT USER_ID, PRODUCT_ID
FROM ONLINE_SALE 
GROUP BY USER_ID, PRODUCT_ID 
HAVING COUNT(SALES_DATE) >= 2
ORDER BY USER_ID ASC, PRODUCT_ID DESC
```

### **<span style="color:#FFAEC9"> 21. 업그레이드 된 아이템 구하기</span>**

### **<span style="color:#FFAEC9"> 27. 대장균들의 자식의 수 구하기</span>**

### 28. 대장균의 크기에 따라 분류하기 1

```python
SELECT 
    ID,
    CASE
        WHEN SIZE_OF_COLONY <= 100 THEN 'LOW'
        WHEN SIZE_OF_COLONY <= 1000 THEN 'MEDIUM'
        ELSE 'HIGH'
    END
FROM ECOLI_DATA
```

### 29. 특정 형질을 가지는 대장균 찾기

```python
SELECT COUNT(ID) AS COUNT
FROM ECOLI_DATA
WHERE 
    (GENOTYPE & 2 <> 2) 
    AND ((GENOTYPE & 1 = 1) OR (GENOTYPE & 4 = 4)) 
```

### 30. 부모의 형질을 모두 가지는 대장균 찾기

```python
SELECT C.ID, C.GENOTYPE, P.GENOTYPE
FROM ECOLI_DATA AS P
JOIN ECOLI_DATA AS C 
ON (P.ID = C.PARENT_ID) AND (P.GENOTYPE & C.GENOTYPE = P.GENOTYPE)
```

### **<span style="color:#FFAEC9">31. 대장균의 크기에 따라 분류하기 2</span>**

```python
-- Searched CASE
SELECT
    ID,
    CASE
        WHEN NTILE(4) OVER (ORDER BY SIZE_OF_COLONY DESC) = 1 THEN 'CRITICAL'
        WHEN NTILE(4) OVER (ORDER BY SIZE_OF_COLONY DESC) = 2 THEN 'HIGH'
        WHEN NTILE(4) OVER (ORDER BY SIZE_OF_COLONY DESC) = 3 THEN 'MEDIUM'
        ELSE 'LOW'
    END AS COLONY_NAME
FROM ECOLI_DATA
ORDER BY ID ASC
```

```python
-- Simple CASE
SELECT
    ID,
    CASE NTILE(4) OVER (ORDER BY SIZE_OF_COLONY DESC)
        WHEN 1 THEN 'CRITICAL'
        WHEN 2 THEN 'HIGH'
        WHEN 3 THEN 'MEDIUM'
        ELSE 'LOW'
    END
FROM ECOLI_DATA
ORDER BY ID ASC
```

### **<span style="color:#FFAEC9">32. 특정 세대의 대장균 찾기</span>**

```python
SELECT A.ID
FROM ECOLI_DATA AS A
LEFT JOIN ECOLI_DATA AS B ON A.PARENT_ID = B.ID
JOIN ECOLI_DATA AS C ON B.PARENT_ID = C.ID
WHERE C.PARENT_ID IS NULL
ORDER BY A.ID ASC
```

### **<span style="color:#FFAEC9">33. 멸종위기의 대장균 찾기</span>**

```python
WITH RECURSIVE GEN AS (
    SELECT ID, PARENT_ID, 1 AS GENERATION
    FROM ECOLI_DATA
    WHERE PARENT_ID IS NULL
    
    UNION ALL
    
    SELECT E.ID, E.PARENT_ID, GEN.GENERATION + 1
    FROM ECOLI_DATA AS E
    JOIN GEN ON E.PARENT_ID = GEN.ID
)

SELECT COUNT(GEN.GENERATION), GEN.GENERATION
FROM ECOLI_DATA AS ED
JOIN GEN ON GEN.ID = ED.ID
WHERE ED.ID NOT IN (
    SELECT PARENT_ID
    FROM ECOLI_DATA
    WHERE PARENT_ID IS NOT NULL
)
GROUP BY GEN.GENERATION
ORDER BY GEN.GENERATION ASC
```
