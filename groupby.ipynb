### 1. 자동차 대여 기록에서 대여중 / 대여 가능 여부 구분하기

```python
SELECT CAR_ID,
    MAX(CASE 
        WHEN START_DATE <= '2022-10-16' AND '2022-10-16' <= END_DATE THEN '대여중'
        ELSE '대여 가능'
    END) AS AVAILABILITY
FROM CAR_RENTAL_COMPANY_RENTAL_HISTORY
GROUP BY CAR_ID
ORDER BY CAR_ID DESC
```

### 2. 저자 별 카테고리 별 매출액 집계하기

```python
SELECT BI.AUTHOR_ID, BI.AUTHOR_NAME, BI.CATEGORY, SUM(S.SALES * BI.PRICE)
FROM BOOK_SALES AS S
JOIN (
    SELECT B.BOOK_ID, B.CATEGORY, B.AUTHOR_ID, A.AUTHOR_NAME, B.PRICE
    FROM BOOK AS B
    JOIN AUTHOR AS A ON A.AUTHOR_ID = B.AUTHOR_ID
    ) AS BI ON S.BOOK_ID = BI.BOOK_ID
WHERE S.SALES_DATE >= '2022-01-01' AND S.SALES_DATE < '2022-02-01'
GROUP BY BI.AUTHOR_ID, BI.CATEGORY
ORDER BY BI.AUTHOR_ID ASC, BI.CATEGORY DESC
```

### 3. 성분으로 구분한 아이스크림 총 주문량

```python
SELECT I.INGREDIENT_TYPE, SUM(F.TOTAL_ORDER)
FROM FIRST_HALF AS F
JOIN ICECREAM_INFO AS I ON F.FLAVOR = I.FLAVOR
GROUP BY I.INGREDIENT_TYPE
ORDER BY SUM(F.TOTAL_ORDER)
```

### 4. 식품분류별 가장 비싼 식품의 정보 조회하기

```python
SELECT F.CATEGORY, M.MAX_PRICE, F.PRODUCT_NAME
FROM FOOD_PRODUCT AS F
JOIN (
    SELECT CATEGORY, MAX(PRICE) AS MAX_PRICE
    FROM FOOD_PRODUCT
    WHERE CATEGORY IN ('과자', '국', '김치', '식용유')
    GROUP BY CATEGORY
    ) AS M ON F.CATEGORY = M.CATEGORY AND M.MAX_PRICE = F.PRICE
ORDER BY M.MAX_PRICE DESC
```

### 5. 진료과별 총 예약 횟수 출력하기

```python
SELECT MCDP_CD AS '진료과 코드', COUNT(PT_NO) AS '5월예약건수'
FROM APPOINTMENT
WHERE APNT_YMD >= '2022-05-01' AND APNT_YMD < '2022-06-01'
GROUP BY MCDP_CD
ORDER BY `5월예약건수` ASC, `진료과 코드` ASC
```

### 6. 즐겨찾기가 가장 많은 식당 정보 출력하기

```python
SELECT R.FOOD_TYPE, R.REST_ID, R.REST_NAME, FM.MAX_F
FROM REST_INFO AS R
JOIN (
    SELECT FOOD_TYPE, MAX(FAVORITES) AS MAX_F
    FROM REST_INFO
    GROUP BY FOOD_TYPE
    ) AS FM 
    ON R.FOOD_TYPE = FM.FOOD_TYPE AND FM.MAX_F = R.FAVORITES
ORDER BY R.FOOD_TYPE DESC
```

### 7. 자동차 종류 별 특정 옵션이 포함된 자동차 수 구하기

```python
SELECT CAR_TYPE, COUNT(CAR_ID) AS CARS
FROM CAR_RENTAL_COMPANY_CAR
WHERE OPTIONS LIKE '%통풍시트%'
    OR OPTIONS LIKE '%열선시트%'
    OR OPTIONS LIKE '%가죽시트%'
GROUP BY CAR_TYPE
ORDER BY CAR_TYPE ASC
```

### 8. 카테고리 별 도서 판매량 집계하기

```python
SELECT B.CATEGORY, SUM(SALES) AS TOTAL_SALES
FROM BOOK AS B
JOIN BOOK_SALES AS BS ON B.BOOK_ID = BS.BOOK_ID
WHERE BS.SALES_DATE >= '2022-01-01' AND BS.SALES_DATE < '2022-02-01'
GROUP BY B.CATEGORY
ORDER BY B.CATEGORY ASC
```
